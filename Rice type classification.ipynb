```python
train = data.take(train_size)
validation = data.skip(train_size).take(validation_size)
test = data.skip(train_size+validation_size).take(test_size)
```

```python
# model 
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=25)  # Stop training if validation loss doesn't improve after 3 epochs

inputs = Input(shape=(256,256,3))
x = Conv2D(16,(3,3),padding='same',activation='relu')(inputs)
x = MaxPooling2D()(x)
x = Conv2D(32,(3,3),activation='relu')(x)
x = Flatten()(x)
outputs = Dense(5,activation='softmax')(x)

model = Model(inputs= inputs,outputs=outputs)

model.compile(optimizer='adam',metrics=['accuracy'],loss='sparse_categorical_crossentropy')

hist = model.fit(train,validation_data=validation,epochs=3,callbacks=[early_stopping])
```
